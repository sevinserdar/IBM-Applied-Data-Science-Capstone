## Tasks

### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
# SQL sorgusunu tanımla
sql_query = """
SELECT DISTINCT "Launch_Site"
FROM SPACEXTABLE;
"""

# Sorguyu çalıştır
cur.execute(sql_query)

# Sonuçları al
launch_sites = cur.fetchall()

# Sonuçları ekrana yazdır
for site in launch_sites:
    print(site[0])
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'CCA'


```python
# SQL sorgusunu tanımla
sql_query = """
SELECT *
FROM SPACEXTABLE
WHERE "Launch_Site" LIKE 'CCA%'
LIMIT 5;
"""

# Sorguyu çalıştır
cur.execute(sql_query)

# Sonuçları al
records = cur.fetchall()

# Sonuçları ekrana yazdır
for record in records:
    print(record)
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
# prompt: Display the total payload mass carried by boosters launched by NASA (CRS)

sql_query = """
SELECT SUM("PAYLOAD_MASS__KG_")
FROM SPACEXTABLE
WHERE "Customer" = 'NASA (CRS)';
"""

# Sorguyu çalıştır
cur.execute(sql_query)

# Sonucu al
total_payload_mass = cur.fetchone()[0]

# Sonucu ekrana yazdır
print(f"Total payload mass carried by boosters launched by NASA (CRS): {total_payload_mass} kg")
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
# prompt: Display average payload mass carried by booster version F9 v1.1

sql_query = """
SELECT AVG("PAYLOAD_MASS__KG_")
FROM SPACEXTABLE
WHERE "Booster_Version" = 'F9 v1.1';
"""

# Sorguyu çalıştır
cur.execute(sql_query)

# Sonucu al
average_payload_mass = cur.fetchone()[0]

# Sonucu ekrana yazdır
print(f"Average payload mass carried by booster version F9 v1.1: {average_payload_mass} kg")
```

### Task 5

##### List the date when the first succesful landing outcome in ground pad was acheived.


_Hint:Use min function_


```python
# prompt: List the date when the first succesful landing outcome in ground pad was acheived.
# Hint:Use min function

sql_query = """
SELECT MIN(Date)
FROM SPACEXTABLE
WHERE "Landing_Outcome" = 'Success (ground pad)';
"""

# Execute the query
cur.execute(sql_query)

# Fetch the result
date = cur.fetchone()[0]

# Print the result
print(f"First successful landing outcome in ground pad: {date}")
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
# prompt: List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000

sql_query = """
SELECT "Booster_Version"
FROM SPACEXTABLE
WHERE "Landing_Outcome" = 'Success (drone ship)'
AND "PAYLOAD_MASS__KG_" BETWEEN 4000 AND 6000;
"""

# Execute the query
cur.execute(sql_query)

# Fetch the results
booster_versions = cur.fetchall()

# Print the results
for booster_version in booster_versions:
    print(booster_version[0])
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
# prompt: List the total number of successful and failure mission outcomes

# Define the SQL query to count successful and failed missions
sql_query = """
SELECT "Landing_Outcome", COUNT(*) AS Total
FROM SPACEXTABLE
GROUP BY "Landing_Outcome";
"""

# Execute the query
cur.execute(sql_query)

# Fetch the results
results = cur.fetchall()

# Print the results
for outcome, count in results:
    print(f"{outcome}: {count}")
```

### Task 8



##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
# prompt: List the names of the booster_versions which have carried the maximum payload mass. Use a subquery

sql_query = """
SELECT "Booster_Version"
FROM SPACEXTABLE
WHERE "PAYLOAD_MASS__KG_" = (SELECT MAX("PAYLOAD_MASS__KG_") FROM SPACEXTABLE);
"""

# Execute the query
cur.execute(sql_query)

# Fetch the results
booster_versions = cur.fetchall()

# Print the results
for booster_version in booster_versions:
    print(booster_version[0])
```

### Task 9


##### List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.

**Note: SQLLite does not support monthnames. So you need to use  substr(Date, 6,2) as month to get the months and substr(Date,0,5)='2015' for year.**


```python
# prompt: List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.
# Note: SQLLite does not support monthnames. So you need to use substr(Date, 6,2) as month to get the months and substr(Date,0,5)='2015' for year.

sql_query = """
SELECT substr(Date, 6,2) AS Month, "Landing_Outcome", "Booster_Version", "Launch_Site"
FROM SPACEXTABLE
WHERE substr(Date,0,5)='2015' AND "Landing_Outcome" LIKE 'Failure (drone ship)%'
"""

# Execute the query
cur.execute(sql_query)

# Fetch the results
records = cur.fetchall()

# Print the results
for record in records:
    print(record)
```

### Task 10




##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order.


```python
# prompt: Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order.

sql_query = """
SELECT "Landing_Outcome", COUNT(*) AS Count
FROM SPACEXTABLE
WHERE Date BETWEEN '2010-06-04' AND '2017-03-20'
GROUP BY "Landing_Outcome"
ORDER BY Count DESC;
"""

# Execute the query
cur.execute(sql_query)

# Fetch the results
results = cur.fetchall()

# Print the results
for outcome, count in results:
    print(f"{outcome}: {count}")
```
